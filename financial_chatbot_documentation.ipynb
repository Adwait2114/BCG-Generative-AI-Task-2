### Financial Inquiry Chatbot: Summary and Documentation

---

### Summary
The Financial Inquiry Chatbot is a command-line based assistant designed to answer financial questions using structured data extracted from an HTML table. The chatbot supports queries regarding revenue, net income, total assets, liabilities, and operating cash flow for companies like Microsoft, Apple, and Tesla across fiscal years 2022 to 2024.

---

### How It Works

1. **Data Source:**

   - Financial data is extracted from an HTML table using `pandas.read_html()` and saved into a CSV file named `financial_data.csv`.

2. **Chatbot Engine:**

   - The chatbot script (`chatbot.py`) loads the CSV using `pandas.read_csv()`.
   - It runs an interactive loop where users can type questions.
   - The script parses the input for company name and fiscal year.
   - Based on keywords like "revenue", "net income", etc., it fetches and displays the appropriate value.

3. **Supported Questions:**

   - What was the revenue of [Company] in [Year]?
   - What is the net income of [Company] in [Year]?
   - What are the total assets of [Company] in [Year]?
   - What are the total liabilities of [Company] in [Year]?
   - What is the operating cash flow of [Company] in [Year]?

4. **Exit:**

   - Users can type `exit` to quit the chatbot.

---

### Example Usage

```
You: What was the revenue of Microsoft in 2023?
Bot: Total Revenue for Microsoft in 2023 was 211.90

You: What is the operating cash flow of Apple in 2022?
Bot: Operating Cash Flow for Apple in 2022 was 122.151

You: exit
Goodbye!
```
